```python
import re
def other_count(document):
    if document.find('[]')!=-1:
        return 0
    # 使用正则表达式匹配出列表信息
    pattern = r"\[('+[\d]+->[\d]+'+(?:, '+[\d]+->[\d]+'+)*)\]"
    match = re.search(pattern, document)
    # 获取匹配到的字符串
    matched_string = match.group(1)
    # 将字符串按逗号和空格分割为列表元素
    elements = matched_string.split(', ')
    # 将元素转换为元组并添加到结果列表
    result = [tuple(map(int, element[1:-1].split('->'))) for element in elements]
    cnt = 0
    for t in result:
        if t[0]!=t[1]:
            cnt+=1
    return cnt
```

# fMRI

```python
import os
import pandas as pd
pd.set_option('display.float_format', '{:.6f}'.format)
results = []
for i in range(1,29):
    task_name = f'fMRI_{i}'
    fname = f'logs/{task_name}.log'
    if os.path.exists(fname):
        with open(fname, 'r') as f: #打开文件
            lines = f.readlines() #读取所有行
            result={
                "Precision'": float(lines[-8].split(':')[-1][:-1]),
                "Recall'": float(lines[-7].split(':')[-1][:-1]),
                "F1'": float(lines[-6].split(':')[-1][:-1]),
                "Precision": float(lines[-4].split(':')[-1][:-1]),
                "Recall": float(lines[-3].split(':')[-1][:-1]),
                "F1": float(lines[-2].split(':')[-1][:-1]),
                "PoD": float(lines[-1].split(':')[-1][:-2])/100
            }
            results.append(result)
df = pd.DataFrame(results, index=[i for i in range(1,len(results)+1)])

print(df)
df.describe()
```
